```python
#importing the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#importing dataset
dataset = pd.read_csv('Position_Salaries.csv')
X = dataset.iloc[:, 1:-1].values
Y = dataset.iloc[:,-1].values
```

```python
#Training the linear regression model on whole dataset
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X, Y)
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree=4)
X_poly = poly_reg.fit_transform(X)
poly_reg.fit(X_poly, Y)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly, Y)
```

```python
#Visualising linear regression results
plt.scatter(X, Y, color='red')
plt.plot(lin_reg.predict(X), color='blue')
plt.title('Truth or Bluff(Linear Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

```python
#Visualising polynomial regression results
plt.scatter(X, Y, color='red')
plt.plot(lin_reg_2.predict(X_poly), color='blue')
plt.title('Truth or Bluff(Polynomial Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

```python
#Visulaising  the polynomial regression with higher resolution and smooth curve
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape(len(X_grid), 1)
plt.scatter(X, Y, color = 'red')
plt.plot(X_grid, lin_reg_2.predict(poly_reg.fit_transform(X_grid)), color = 'blue')
plt.title('Truth or Bluff(Polynomial Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

```python
#Prediciting a new result with linear regression
lin_reg.predict([[6.5]])
```

```python
#prediciting a new result with polynomial regression
lin_reg_2.predict(poly_reg.fit_transform([[6.5]]))
```
